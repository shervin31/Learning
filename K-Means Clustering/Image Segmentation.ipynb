# Image Segmentation 

### In this module I preprocessed a brain scan using k-means clustering algorithm to prepare it for tumour classification. 

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
image = mpl.image.imread("Brain Tumour.png")
plt.imshow(image)
plt.axis("off")
```

```python
## White, red blue, black
image.shape
```

```python
image_rgb = image

X = image_rgb.reshape(-1, 4)
```

```python
kmeans = KMeans(n_clusters=3, n_init=10, random_state=0)
```

```python
kmeans.fit(X)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image_rgb.shape)
```

```python
plt.imshow(segmented_img)
plt.axis("off")
plt.show()
```
