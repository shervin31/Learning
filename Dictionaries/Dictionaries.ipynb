### Dictionary

```python
## Key: Value 
x = {
    "name": "David", 
    "age": 10
}
```

### Nested Dictionary

```python
y = {
    "Shervin": {"Math": 95, "English": 92},
    "David": {"Math": 91, "English": 85}
}
```

### Indexing Dictionaries 
##### How do you access the english mark of Shervin? 

```python
## We must acess the key (Shervin), then the english field. 

print(y['Shervin']['English'])
```

### Updating Dictionary 

```python
y['Shervin']['Math'] = 99
print(y['Shervin']['Math'])
```

### Adding new value 

```python
y['Alex'] = {
    'Math': "90",
    'Eglish': '85'
}
print(y) 
```

### Dictionary Methods 

```python
## Getting keys
print(y.keys())

## Getting Values 
print(y.values())

## Getting key value pairs 
print(y.items())

## Clearing dictionary
# y.clear()

## get() lets you pass a key and it gets corresponding values
print(y.get("Shervin"))

## Delete item by passing key
y.pop("Alex")
print(y)
```
