### Escape Characters 

- \n is newline  
- \t is tab (8 spaces)
- \' is single quote 
- \'' is double quote

### String Formatting 

- %s is for strings 
- %d is for integers 
- %f is for float 
- %. is number of digits after decimal 

### New String Formatting (Python version 3)

- Use {} for evertything and puut .format 
- For decimal places, do (:.x)

```python
first_name = 'Shervin'
last_name = 'Zare'
age = 20.7423
new_formatted_string = 'My name is {} {} and I am {:.0f} years old'.format(first_name, last_name, int(age))
print(new_formatted_string)
```

### String Methods 


```python
greeting = "hello world"

# Reversing a String 
print(greeting[::-1])

# Skipping Characters 
print(greeting[0::2])

# .capitalize() capitalzies first letter 
print(greeting.capitalize())

# .count() counts number of substrings
# You can do between start and end
print(greeting.count('l'))
print(greeting.count('l', 0, 5))

# .endswith() 
print(greeting.endswith('d'))

# .find() returns index of the first occurence of substring, otherwise -1 
print(greeting.find('wor'))

# .rfind() returns index of the last occurence of substring, otherwise -1 
print(greeting.rfind('l'))

# .replace() replaces strings with new one 
print(greeting.replace('world', 'everyone'))
```
